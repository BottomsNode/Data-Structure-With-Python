# <center>Simple Queue and Circular Queue Animation Python Code</center>

```python
import tkinter as tk
from tkinter import messagebox

class QueueAnimation:
    def __init__(self, root):
        self.root = root
        self.root.title("Queue Animation by Nishit Shivdasani")
        
        self.canvas_width = 800
        self.canvas_height = 400
        self.queue_width = 40
        self.queue_height = 40
        self.queue_spacing = 10
        self.queue_size = 0
        self.queue_data = []
        
        self.queue_type = tk.StringVar()
        self.queue_type.set("Simple")  # Default queue type is Simple
        
        self.canvas = tk.Canvas(root, width=self.canvas_width, height=self.canvas_height, bg="white")
        self.canvas.pack(pady=10)
        
        self.label_size = tk.Label(root, text="Queue Size:", font=("Helvetica", 12))
        self.label_size.pack()
        self.entry_size = tk.Entry(root, width=10)
        self.entry_size.pack()
        
        self.label_data = tk.Label(root, text="Data to Enqueue:", font=("Helvetica", 12))
        self.label_data.pack()
        self.entry_data = tk.Entry(root, width=20)
        self.entry_data.pack()
        
        self.radio_simple = tk.Radiobutton(root, text="Simple Queue", variable=self.queue_type, value="Simple", font=("Helvetica", 12))
        self.radio_simple.pack()
        
        self.radio_circular = tk.Radiobutton(root, text="Circular Queue", variable=self.queue_type, value="Circular", font=("Helvetica", 12))
        self.radio_circular.pack()
        
        self.button_enqueue = tk.Button(root, text="Enqueue", command=self.enqueue, font=("Helvetica", 12))
        self.button_enqueue.pack(pady=5)
        
        self.button_dequeue = tk.Button(root, text="Dequeue", command=self.dequeue, font=("Helvetica", 12))
        self.button_dequeue.pack(pady=5)
        
        self.button_display = tk.Button(root, text="Display", command=self.display_queue, font=("Helvetica", 12))
        self.button_display.pack(pady=5)
        
        self.owner_label = tk.Label(root, text="Developed by Nishit Shivdasani", font=("Helvetica", 10), fg="gray")
        self.owner_label.pack(side="bottom", pady=10)
    
    def enqueue(self):
        data = self.entry_data.get()
        if len(self.queue_data) < int(self.entry_size.get()):
            if self.queue_type.get() == "Simple":
                self.queue_data.append(data)
            elif self.queue_type.get() == "Circular":
                self.queue_data.insert(0, data)  # Inserting at the front for circular queue
            self.queue_size += 1
            self.draw_queue()
        else:
            tk.messagebox.showerror("Error", "Queue Overflow!")
    
    def dequeue(self):
        if len(self.queue_data) > 0:
            if self.queue_type.get() == "Simple":
                self.queue_data.pop(0)
            elif self.queue_type.get() == "Circular":
                self.queue_data.pop()
            self.queue_size -= 1
            self.canvas.delete("all")
            self.draw_queue()
        else:
            tk.messagebox.showerror("Error", "Queue Underflow!")
    
    def display_queue(self):
        tk.messagebox.showinfo("Queue", "Queue Contents: " + ", ".join(self.queue_data))
    
    def draw_queue(self):
        self.canvas.delete("all")
        x_start = (self.canvas_width - (self.queue_size * (self.queue_width + self.queue_spacing))) / 2
        y = self.canvas_height / 2 - self.queue_height / 2
        
        for i in range(self.queue_size):
            x = x_start + i * (self.queue_width + self.queue_spacing)
            self.canvas.create_rectangle(x, y, x + self.queue_width, y + self.queue_height, fill="lightblue", outline="black")
            self.canvas.create_text(x + self.queue_width / 2, y + self.queue_height / 2, text=str(self.queue_data[i]), font=("Arial", 12, "bold"))

def main():
    root = tk.Tk()
    app = QueueAnimation(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```
