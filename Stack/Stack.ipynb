# <center>Stack Python Code</center>

```python
import tkinter as tk

class Stack:
    def __init__(self):
        self.stack = []

    def is_empty(self):
        return len(self.stack) == 0

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if not self.is_empty():
            return self.stack.pop()
        else:
            return None

class StackApp:
    def __init__(self, master):
        self.master = master
        self.master.title("Stack by Nishit Shivdasani")
        
        self.master.geometry("400x300")
        
        self.stack = Stack()

        self.label_operation = tk.Label(master, text="Operation:", font=("Helvetica", 12))
        self.label_operation.pack(pady=5)

        self.entry_operation = tk.Entry(master, width=20)
        self.entry_operation.pack(pady=5)

        self.push_button = tk.Button(master, text="Push", command=self.push_element, font=("Helvetica", 12))
        self.push_button.pack(pady=5)

        self.pop_button = tk.Button(master, text="Pop", command=self.pop_element, font=("Helvetica", 12))
        self.pop_button.pack(pady=5)

        self.display_button = tk.Button(master, text="Display", command=self.display_stack, font=("Helvetica", 12))
        self.display_button.pack(pady=5)

        self.output_label = tk.Label(master, text="", font=("Helvetica", 12), fg="blue")
        self.output_label.pack(pady=10)
        
        self.owner_label = tk.Label(master, text="Developed by Nishit Shivdasani", font=("Helvetica", 10), fg="gray")
        self.owner_label.pack(side="bottom", pady=10)

    def push_element(self):
        operation = self.entry_operation.get()
        if operation:
            self.stack.push(operation)
            self.output_label.config(text="Element pushed: {}".format(operation))
        else:
            self.output_label.config(text="Please enter an element.")

    def pop_element(self):
        popped_item = self.stack.pop()
        if popped_item is not None:
            self.output_label.config(text="Element popped: {}".format(popped_item))
        else:
            self.output_label.config(text="Stack is empty.")

    def display_stack(self):
        if not self.stack.is_empty():
            output = "Stack: " + ", ".join(str(item) for item in self.stack.stack[::-1])
            self.output_label.config(text=output)
        else:
            self.output_label.config(text="Stack is empty.")

if __name__ == "__main__":
    root = tk.Tk()
    app = StackApp(root)
    root.mainloop()
```
